```python
import pyarrow.parquet as pq
from sqlalchemy import create_engine
from tqdm.auto import tqdm

file = '/Users/iljasmarjov/Downloads/green_tripdata_2025-11.parquet'

engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')

# Open the parquet file
parquet_file = pq.ParquetFile(file)
# parquet_file.read()
# # Iterate through row groups (natural chunks in Parquet)
# for batch in parquet_file.iter_batches(batch_size=100000):
#     df = batch.to_pandas()
#     df.to_sql('green_taxi_trips', engine, if_exists='append', index=False)

first = True

for batch in parquet_file.iter_batches(batch_size=100000):

    if first:
        # Create table schema (no data)
        batch.head(0).to_sql(
            name="green_taxi_data",
            con=engine,
            if_exists="replace"
        )
        first = False
        print("Table created")

    # Insert chunk
    batch.to_sql(
        name="green_taxi_trips",
        con=engine,
        if_exists="append"
    )

    print("Inserted:", len(batch))
```
